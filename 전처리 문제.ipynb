### 공기의 질을 측정한 데이터
1. airquanlity 데이터에 대해서 결측치가 가장 많은 변수를 찾아서 해당 결측치를 0으로 대체하고 결측치를 제외한 평균과 0으로 대치한 후 평균의 차이를 구하시오
2. wind 변수에 대해서 Min-Max 정규화를 수행 한 후 평균 값과 Z 정규화를 수행 한 후 평균 값의 차이를 구하시오
3. 월별 평균 기온을 구하시오

```python
import pandas as pd

air=pd.read_csv("./datasets/airquality.csv")
air.isna().sum()   ### ozone
```

```python
M1=air['Ozone'].mean()
```

```python
M2=air['Ozone'].fillna(0).mean()
```

```python
### 1번 답
print(M1-M2)  
```

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler

MmScaler=MinMaxScaler()
StdScaler = StandardScaler()

MmScaler.fit(air['Wind'].values.reshape(-1,1))
StdScaler.fit(air['Wind'].values.reshape(-1,1))

W1=MmScaler.transform(air['Wind'].values.reshape(-1,1)).mean()
W2=StdScaler.transform(air['Wind'].values.reshape(-1,1)).mean()

### 2번 문제 답
print(W1-W2)   
```

```python
### 3번 문제 답
air.groupby(['Month']).mean()[['Temp']]
```

### 근무자 정보를 측정한 데이터
1. Attrition은 퇴사 여부 데이터를 (yes:1, no:0)으로 데이터를 변환하여 새로운 파생변수에 대입하고 범주별 레코드 수를 계산하시오
2. 범주형 변수 중 유일한 값을 1개만 가지고 있는 컬럼을 찾아내어 그 컬럼을 데이터에서 제거하시오
3. 원래 데이터에서 숫자형인 컬럼만 추출하여 새로운 데이터프레임으로 생성하고, 각 변수간의 상관계수(피어슨 상관계수)를 구하시오

```python
HR_E=pd.read_csv("./datasets/HR-Employee-Attrition.csv")
HR_E['Attrition'].value_counts()
```

```python
HR_E['Attrition_2']=HR_E['Attrition'].map(
    {
        'Yes': 1,
        'No' : 0 
    }
)
```

```python
## 새로운 파생변수 Attrition_2에 저장후, 범주별 레코드 수 0:1233,1:237로 계산
HR_E['Attrition_2'].value_counts()
```

```python
HR_E['EmployeeCount'].value_counts()
```

```python
## 2번 제거
HR_E.drop('EmployeeCount',axis=1,inplace=True)
```

```python
HR_E.info()
```

```python
import numpy as np

int_list=[]
for col in HR_E.columns:
   
    if HR_E[col].dtype == np.int64:
        int_list.append(col)
print(int_list)
```

```python
HR_E[int_list].info()
```

```python
## 3번 상관관계 구했습니다
HR_E[int_list].corr()
```
