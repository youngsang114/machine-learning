```python
iris_df=pd.DataFrame(columns=iris.feature_names , data=iris.data)
iris_df['label']=iris.target
```

# 2.데이터 정제

```python
iris_df.isnull().sum()
```

```python
iris_df.duplicated().sum()
```

```python
iris_df[iris_df.duplicated()]
```

```python
iris_df.drop_duplicates(keep='first',inplace=True)
iris_df.info()
```

```python
# 데이터 분할
X=iris_df.iloc[:,:4]
y=iris_df['label']

X_train,X_test,y_train,y_test =train_test_split(X,y,random_state=42,stratify=y)
```

```python
print("학습용 데이터 수:{0},테스트용 데이터 수 :{1}".format(len(X_train),len(X_test)))
```

```python
#DecisionTreeClassifier 객체 생성
model=DecisionTreeClassifier(random_state=42)
model.fit(X_train,y_train,sample_weight=None)
```

```python
from sklearn.tree import export_graphviz
from subprocess import call
from IPython.display import Image


export_graphviz(model, out_file='tree.dot')
call(['dot', '-Tpng', 'tree.dot', '-o', 'tree.png', '-Gdpi=600'])


Image(filename='tree.png')
```

```python
# 테스트용 데이터로 모델 평가
score=model.score(X_test,y_test)
score
```

```python
new=np.array([[5.4,4,1.5,0.2]])
y_pred=model.predict(new)
y_pred
```
