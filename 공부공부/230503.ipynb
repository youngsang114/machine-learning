### 교차검증(Cross validation)
    - K-fold
    - Stratified K-fold
    

```python
import pandas as pd
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
iris=load_iris()
iris_df=pd.DataFrame(columns=iris.feature_names , data=iris.data)
iris_df['label']=iris.target
```

```python

X=iris_df.iloc[:,:4]
y=iris_df['label']

X_train,X_test,y_train,y_test =train_test_split(X,y,random_state=42 ,stratify=y)
#DecisionTreeClassifier 객체 생성
model=DecisionTreeClassifier(random_state=42)
model.fit(X_train,y_train,sample_weight=None)
```

```python
# k-fold
from sklearn.model_selection import KFold

kf=KFold(n_splits=3,shuffle=False)
for train_index ,text_index in kf.split(X):
    print("----------------------------------")
    print("학습용:", train_index)
    print("\n학습용 레이블:",y[train_index].unique())
    print("\n테스트용:",text_index)
    print("\n테스트용 레이블:",y[text_index].unique())
```

```python
# StratifiedKFold

from sklearn.model_selection import StratifiedKFold

skf= StratifiedKFold(n_splits=3)
for train_index,test_index in skf.split(X,y):
    print('-------------------------------')
    print('학습용:',train_index)
    print("\n 학습용 레이블:",y[train_index].unique())
    print('테스트용:',test_index)
    print("\n 테스트트용 레이블:",y[test_index].unique())
```

```python
# cross_validate()

from sklearn.model_selection import cross_validate

scores=cross_validate(model, X_train, y_train , cv=3, return_estimator=True)
scores
```

```python
# 3개의 분류기 평가

for i in range(3):
    score=scores['estimator'][i].score(X_test,y_test)
    print('{0}번째 의사결정나무 정확도:{1:}'.format(i+1,score))
```
